## Problem 1
## Sampling and Observing Normal Distributions
In this problem, we will:

1- Characterize the Distribution: Analyze the distribution of the sample means by calculating its standard deviation.

2- Examine Spread with Increasing Sample Size: Observe how the spread of the distribution decreases with increasing sample size. This aligns with the intuitive notion that larger samples tend to have means closer to the true mean of the underlying population from which the samples are drawn.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Different sample sizes we are going to try
sample_sizes = [3, 10, 50, 100, 500, 1000]
```

```python
# We will use the list below to keep the SD of the distribution of the means at each given sample size
# (note that it's ok to initialize with an empty list - it will dynamically grow as we append results)
mean_sds = []
```

```python
# Number of samples we want to draw for each sample size
num_samples = 1000
```

```python
# Loop through each sample size
for N in sample_sizes:
    # List to store the means of the samples
    m = []
    # Draw num_samples samples of size N
    for _ in range(num_samples):
        # Draw a sample of size N from a normal distribution (mean=0, std=1)
        sample = np.random.normal(loc=0, scale=1, size=N)
        # Calculate the mean of the sample and append it to the list m
        m.append(np.mean(sample))
        # Calculate the standard deviation of the means in list m
    sd_of_means = np.std(m)
    # Append the standard deviation to the list mean_sds
    mean_sds.append(sd_of_means)
```

```python
# Plotting the result
plt.figure(figsize=(10, 6))
plt.plot(sample_sizes, mean_sds, 'o-', label='SEM')
plt.plot(sample_sizes, 1/np.sqrt(sample_sizes), 'b--', label='Theoretical SEM (1/âˆšN)')
plt.xlabel('Sample Size')
plt.ylabel('SEM')
plt.title('SEM vs Sample Size')
plt.legend()
plt.grid(True)
plt.show()
```

***Interpretation***:

The graph shows a decreasing curve, illustrating that the SEM decreases rapidly at first (with smaller sample sizes) and more slowly as N increases. This curve is characteristic of a relationship of the form :*$$ y = \frac{1}{\sqrt{N}} $$*

For instance, when :
*   N = 3, the SEM is relatively high. With small sample sizes, there is greater
variability in the sample means, indicating more uncertainty in estimating the population mean.

*  N = 10, the SEM decreases noticeably, indicating reduced uncertainty with a slightly larger sample.

*  N = 50 and N = 100, the SEM continues to decrease, showing more moderate reductions. This reflects the principle of diminishing returns: each additional increase in sample size reduces uncertainty, but to a lesser extent.

*  N = 500 and N = 1000, the SEM is very low, meaning the estimate of the population mean is very precise. The fluctuations in the sample means are minimal.

This means that to obtain an accurate estimate of the population mean, it is advantageous to use a larger sample size.

## Problem 2
## Sampling and Observing Distribution Changes




**Uniform Distributions**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define sample sizes to try
sample_sizes = [1, 2, 5, 10, 50, 100]

# Number of experiments (number of times we sample)
num_experiments = 1000

# Function to plot histograms of sample means
def plot_sample_means(sample_sizes, num_experiments):
    plt.figure(figsize=(8, 5))

    for i, N in enumerate(sample_sizes, 1):
        # List to store means of samples
        sample_means = [np.mean(np.random.uniform(low=0.0, high=1.0, size=N)) for _ in range(num_experiments)]

        # Plot the histogram of the sample means
        plt.subplot(2, 3, i)
        plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='blue', edgecolor='black')
        plt.title(f'N = {N}')
        plt.xlabel('Sample Mean')
        plt.ylabel('Density')

    plt.tight_layout()
    plt.show()

# Function to plot SEM as a function of sample size
def plot_sem_vs_sample_size(sample_sizes, num_experiments):
    mean_sds = []

    for N in sample_sizes:
        sample_means = [np.mean(np.random.uniform(low=0.0, high=1.0, size=N)) for _ in range(num_experiments)]
        sem = np.std(sample_means)
        mean_sds.append(sem)

    plt.figure(figsize=(8, 5))
    plt.plot(sample_sizes, mean_sds, marker='o')
    plt.xlabel('Sample Size (N)')
    plt.ylabel('Standard Error of the Mean (SEM)')
    plt.title('SEM as a Function of Sample Size')
    plt.grid(True)
    plt.show()

# Plot histograms of sample means for different sample sizes
plot_sample_means(sample_sizes, num_experiments)

# Plot SEM as a function of sample size
plot_sem_vs_sample_size(sample_sizes, num_experiments)
```

**Interpretation**
* Histograms of Sample Means:

The histograms show that for small sample sizes (N), the distribution of sample means is often irregular and may exhibit asymmetry.
As the sample size increases, the distribution of sample means tends to approach a more symmetric shape, although some variability can still be observed for larger sample sizes.

* Standard Error of the Mean (SEM):

The SEM plot as a function of sample size demonstrates that the standard error of the mean decreases as the sample size increases, in accordance with the Central Limit Theorem.
For smaller sample sizes, SEM is generally higher, indicating greater variability in sample means.
Even for larger sample sizes like 50 and 100, the SEM can remain relatively high compared to distributions that are more concentrated.

**Exponential Distributions**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define sample sizes to try
sample_sizes = [1, 2, 5, 10, 50, 100]

# Number of experiments (number of times we sample)
num_experiments = 1000

# Function to plot histograms of sample means
def plot_sample_means(sample_sizes, num_experiments):
    plt.figure(figsize=(8, 5))

    for i, N in enumerate(sample_sizes, 1):
        # List to store means of samples
        sample_means = [np.mean(np.random.exponential(scale=1.0, size=N)) for _ in range(num_experiments)]

        # Plot the histogram of the sample means
        plt.subplot(2, 3, i)
        plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='blue', edgecolor='black')
        plt.title(f'N = {N}')
        plt.xlabel('Sample Mean')
        plt.ylabel('Density')

    plt.tight_layout()
    plt.show()

# Function to plot SEM as a function of sample size
def plot_sem_vs_sample_size(sample_sizes, num_experiments):
    mean_sds = []

    for N in sample_sizes:
        sample_means = [np.mean(np.random.exponential(scale=1.0, size=N)) for _ in range(num_experiments)]
        sem = np.std(sample_means)
        mean_sds.append(sem)

    plt.figure(figsize=(8, 5))
    plt.plot(sample_sizes, mean_sds, marker='o')
    plt.xlabel('Sample Size (N)')
    plt.ylabel('Standard Error of the Mean (SEM)')
    plt.title('SEM as a function of Sample Size')
    plt.grid(True)
    plt.show()

# Plot histograms of sample means for different sample sizes
plot_sample_means(sample_sizes, num_experiments)

# Plot SEM as a function of sample size
plot_sem_vs_sample_size(sample_sizes, num_experiments)
```

**Interpretation**
* Histograms of Sample Means:

For the exponential distribution, histograms also show a trend towards a more normal distribution of sample means as sample size increases.
Sample means for smaller sample sizes can be highly variable and exhibit asymmetry, but they become more symmetric and centered around a mean value as N increases.
* Standard Error of the Mean (SEM):

Similar to the uniform distribution, the SEM plot for the exponential distribution shows a significant decrease in standard error of the mean with increasing sample size.
This indicates that sample means become more predictable and closer to the true population mean as N becomes larger.
